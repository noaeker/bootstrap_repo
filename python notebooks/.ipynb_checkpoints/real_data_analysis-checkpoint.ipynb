```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot(data, bootstrap_cols):
    first_studies = data['study_name'].unique()[:20]
    fig, axes = plt.subplots(4, 5, figsize=(15, 12))

    data =  pd.melt(data[bootstrap_cols+['predicted_bootstrap_score_nni','study_name']], id_vars=['predicted_bootstrap_score_nni', 'study_name'], 
                         value_vars=bootstrap_cols, 
                         var_name='type', value_name='value')
    method_mapping = {'aLRT_iqtree':'aLRT score','aBayes_iqtree':'aBayes score'}
    data['type'] = data['type'].map(method_mapping).fillna(data['type'])



    for idx, study in enumerate(first_studies):
        real_data_small = data.loc[data.study_name == study]
        i = idx // 5
        j = idx % 5    
        ax = axes[i, j]
        sns.scatterplot(x='predicted_bootstrap_score_nni', y='value', data=real_data_small, hue="type", ax=ax, alpha = 0.5)
        ax.plot([0, 1], [0, 1], color='red', linestyle='--', label='y = x')
        ax.set_title(study)
        ax.set_xlim(-0.05, 1.05)  # Set the x-axis limit to start at 0
        ax.set_ylim(-0.05, 1.05)  # Set the y-axis limit to start at 0
        ax.legend().remove()  # Remove the legend
        ax.set_xlabel('Machine-learning based bootstrap score')
        ax.set_ylabel('Comparison bootstrap score')
        plt.tight_layout()

    handles, labels = ax.get_legend_handles_labels()
    fig.legend(handles, labels, loc='upper center',bbox_to_anchor=(0.5, 1.05), ncol=2)
    plt.tight_layout()
    plt.show()
```

```python
plot(iqtree_real_data,['aLRT_iqtree','aBayes_iqtree'])
```

```python
rpl16b = iqtree_real_data_doro.loc[iqtree_real_data_doro.study_name=='rpl16b']
#rpl16b_data =  pd.melt(rpl16b[['aLRT_iqtree','aBayes_iqtree']+['predicted_bootstrap_score_nni','study_name']], id_vars=['predicted_bootstrap_score_nni', 'study_name'], 
#                         value_vars=['aLRT_iqtree','aBayes_iqtree'], 
#                         var_name='type', value_name='value')

rpl16b_data = pd.melt(rpl16b[['aLRT_iqtree', 'aBayes_iqtree', 'predicted_bootstrap_score_nni', 'study_name', 'node_name']],
                      id_vars=['predicted_bootstrap_score_nni', 'study_name', 'node_name'],
                      value_vars=['aLRT_iqtree', 'aBayes_iqtree'],
                      var_name='type', value_name='value')
method_mapping = {'aLRT_iqtree':'aLRT score','aBayes_iqtree':'aBayes score'}
rpl16b_data['type'] = rpl16b_data['type'].map(method_mapping).fillna(rpl16b_data['type'])

rpl16b_data["is_special_node"] = rpl16b_data["node_name"].isin(["N108","N66","N143"])
node_name_mapping = {'N108': 'N1', 'N66': 'N2', 'N143': 'N3'}

# Apply the mapping to the 'node_name' column
rpl16b_data['node_name'] = rpl16b_data['node_name'].map(node_name_mapping).fillna(rpl16b_data['node_name'])


```

```python
print(np.corrcoef(rpl16b["predicted_bootstrap_score_nni"], rpl16b["aLRT_iqtree"] )[0,1]**2)
print(np.corrcoef(rpl16b["predicted_bootstrap_score_nni"], rpl16b["aBayes_iqtree"] )[0,1]**2)
print(np.mean(rpl16b["predicted_bootstrap_score_nni"]))
print(np.mean(rpl16b["aLRT_iqtree"]))
print(np.mean(rpl16b["aBayes_iqtree"]))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots()

# Scatter plot for non-special nodes
sns.scatterplot(x='predicted_bootstrap_score_nni', y='value', data=rpl16b_data[~rpl16b_data['is_special_node']],
                hue="type", ax=ax, alpha=0.5)

# Scatter plot for special nodes, with a different marker shape
sns.scatterplot(x='predicted_bootstrap_score_nni', y='value', data=rpl16b_data[rpl16b_data['is_special_node']],
                hue="type", ax=ax, alpha=0.5, marker='x', s=100, legend=False)  # Use 'x' marker and increase size

# Annotate special nodes with corresponding values from the 'node_name' column
special_nodes_data = rpl16b_data[rpl16b_data['is_special_node']]
for index, row in special_nodes_data.iterrows():
    ax.annotate(row['node_name'], (row['predicted_bootstrap_score_nni'], row['value']),
                textcoords="offset points", xytext=(5,5), ha='center')

# Plotting the diagonal line
ax.plot([0, 1], [0, 1], color='red', linestyle='--', label='y = x')

ax.set_xlim(-0.05, 1.05)
ax.set_ylim(-0.05, 1.05)
ax.legend().remove()

# Create legend
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=2)

ax.set_xlabel('Machine-learning based bootstrap score')
ax.set_ylabel('Comparison bootstrap score')
plt.tight_layout()
plt.savefig(f"/Users/noa/Desktop/Revision_figures/Ecker.165.fig.4.pdf", format="pdf", bbox_inches="tight")
```
