```python
import pandas as pd
from sklearn.calibration import CalibratedClassifierCV, calibration_curve
from matplotlib import pyplot as plt
```

```python
def calibration_plot(pred_dict,true):
    fig, ax = plt.subplots()
    for name in pred_dict:
        pred = pred_dict[name]
        fop_calibrated, mpv_calibrated = calibration_curve(true, pred, n_bins=30)
        # plot perfectly calibrated
        # plot model reliabilities
        ax.plot(fop_calibrated, mpv_calibrated, marker='.', label = name)
    ax.plot([0, 1], [0, 1], linestyle='--', color='black')
    ax.legend()
    pyplot.show()
```

```python
iqtree_test_data = pd.read_csv("/Users/noa/Workspace/bootstrap_results/remote_results/ML_results/iqtree/final_model/test.tsv",sep='\t')
raxml_test_data = pd.read_csv("/Users/noa/Workspace/bootstrap_results/remote_results/ML_results/raxml/final_model/test.tsv",sep='\t')
fasttree_test_data = pd.read_csv("/Users/noa/Workspace/bootstrap_results/remote_results/ML_results/raxml/final_model/test.tsv",sep='\t')
```

```python
calibration_plot({'IQTREE aLRT':iqtree_test_data['predictions_feature_aLRT_iqtree_support_raw_boot'],
                  'IQTREE Ultrafast':iqtree_test_data['predictions_bootstrap_support_raw_boot'],
                 'IQTREE aBayes':iqtree_test_data['predictions_feature_aBayes_iqtree_support_raw_boot'],
                  'IQTREE ML model':iqtree_test_data['predictions_full_standard']
                 }, iqtree_test_data["true_binary_support"])


calibration_plot({'RAxML bootstrap':raxml_test_data['predictions_bootstrap_support_raw_boot'],
                  'RAxML ML model':raxml_test_data['predictions_full_standard']
                 }, iqtree_test_data["true_binary_support"])


calibration_plot({
                  'Fasttree ML model':fasttre_test_data['predictions_full_standard']
                 }, iqtree_test_data["true_binary_support"])




```
